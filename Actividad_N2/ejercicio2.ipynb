Actividad en laboratorio (Procesando imagenes) 

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 6]

img_bgr = cv2.imread('image1.png', cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
cv2.imwrite('image_gray.png', img_gray)
cv2.imwrite('image_bgr.png', img_bgr)

plt.figure(figsize=(12, 6))
plt.subplot(231)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.subplot(232)
plt.imshow(img_gray, cmap='gray')
plt.title('Escala de Grises')
plt.subplot(233)
plt.imshow(img_bgr[:, :, 0], cmap='gray')
plt.title('gris 1')
plt.subplot(234)
plt.imshow(img_bgr[:, :, 1], cmap='gray')
plt.title('gris 2')
plt.subplot(235)
plt.imshow(img_bgr[:, :, 2], cmap='gray')
plt.title('gris 3')
plt.subplot(236)
plt.imshow(img_gray, cmap='jet')
plt.title('Falso color')
plt.show()

```

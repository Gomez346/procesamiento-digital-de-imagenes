```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

colormaps = [
    cv2.COLORMAP_AUTUMN, cv2.COLORMAP_BONE, cv2.COLORMAP_JET, cv2.COLORMAP_WINTER,
    cv2.COLORMAP_RAINBOW, cv2.COLORMAP_OCEAN, cv2.COLORMAP_SUMMER, cv2.COLORMAP_SPRING,
    cv2.COLORMAP_COOL, cv2.COLORMAP_HSV, cv2.COLORMAP_PINK, cv2.COLORMAP_HOT,
    cv2.COLORMAP_PARULA, cv2.COLORMAP_MAGMA, cv2.COLORMAP_INFERNO, cv2.COLORMAP_PLASMA,
    cv2.COLORMAP_VIRIDIS, cv2.COLORMAP_CIVIDIS, cv2.COLORMAP_TWILIGHT, cv2.COLORMAP_TWILIGHT_SHIFTED,
    cv2.COLORMAP_TURBO, cv2.COLORMAP_DEEPGREEN
]


def falso_color(image_gray,colormap):
    escalar_imagen= cv2.normalize(image_gray, None,0,255, cv2.NORM_MINMAX)
    apliocolor= cv2.applyColorMap(escalar_imagen,colormap)
    return apliocolor

image_gray= cv2.imread('img.jpg',0)
fig, axes = plt.subplots(6, 4, figsize=(15, 20))  
fig.suptitle('Colormaps en OpenCV', fontsize=16)

for i, colormap in enumerate(colormaps):
    img_falso = falso_color(image_gray, colormap)
    ax = axes[i // 4, i % 4]
    ax.imshow(cv2.cvtColor(img_falso, cv2.COLOR_BGR2RGB))
    ax.axis('off')
    ax.set_title(f'Color {i + 1}')

plt.tight_layout()
plt.show()
```

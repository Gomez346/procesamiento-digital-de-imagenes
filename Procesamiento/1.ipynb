```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def posterizar(imagen,nivel):
    factor=255/nivel
    imagen_post= np.round((imagen/factor)*factor)
    return imagen_post.astype(np.uint8)

imagen = cv2.imread('img.jpg')
img_gris= cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
nivel=2
nueva= posterizar(img_gris,nivel)


plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(nueva,cmap="gray")
```

convoluci√≥n con un kernel diagonal

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolucion_diagonal(imagen, tamanioKernel):
    kernel = np.eye(tamanioKernel, dtype=np.float32) / tamanioKernel
    imgFin = cv2.filter2D(imagen, -1, kernel)
    return imgFin

imagenes = ['img1.jpeg', 'img2.jpeg', 'img3.jpeg']
tamanioKernel = 5


for i, nombre_imagen in enumerate(imagenes):
    
    imagen = cv2.imread(nombre_imagen)
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    img_res = convolucion_diagonal(gris, tamanioKernel)
    
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.title(f"Original - {nombre_imagen}")
    plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
    plt.axis("off")

    plt.subplot(122)
    plt.title(f"Convoluci√≥n Diagonal - {nombre_imagen}")
    plt.imshow(img_res, cmap='gray')
    plt.axis("off")

    plt.show()
```

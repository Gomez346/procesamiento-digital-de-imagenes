Realiza una investigación científica sobre el filtro Bayer aplicado en el procesamiento digital de Imágenes, el porque se aplica y como proviene;

de la misma manera realizando código lista las versiones de filtro Bayer en cv2 y aplícalos en tus imágenes seleccionadas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("imagen_bayer.jpg", cv2.IMREAD_GRAYSCALE)

filters = {
    "BayerBG2BGR": cv2.COLOR_BayerBG2BGR,
    "BayerGB2BGR": cv2.COLOR_BayerGB2BGR,
    "BayerRG2BGR": cv2.COLOR_BayerRG2BGR,
    "BayerGR2BGR": cv2.COLOR_BayerGR2BGR
}

fig, axes = plt.subplots(1, len(filters)+1, figsize=(15, 5))

axes[0].imshow(image, cmap="gray")
axes[0].set_title("Imagen Bayer")
axes[0].axis("off")

for i, (name, code) in enumerate(filters.items(), start=1):
    bgr_image = cv2.cvtColor(image, code)
    rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)  
    axes[i].imshow(rgb_image)
    axes[i].set_title(name)
    axes[i].axis("off")

plt.show()
```

Aplica las las transformaciones de YUV2RGB, RGB2YUVy viceversa a la imagen seleccionada

```python
import cv2
import matplotlib.pyplot as plt


image = cv2.imread("image1.png")  
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 


image_yuv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2YUV)


image_rgb_converted = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2RGB)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Imagen a RGB
axes[0].imshow(image_rgb)
axes[0].set_title("Imagen Original (RGB)")
axes[0].axis("off")

# Imagen convertida a YUV
axes[1].imshow(image_yuv)
axes[1].set_title("Imagen en YUV")
axes[1].axis("off")

# Imagen convertida de YUV a RGB
axes[2].imshow(image_rgb_converted)
axes[2].set_title("Imagen convertida de YUV a RGB")
axes[2].axis("off")

plt.show()
```
